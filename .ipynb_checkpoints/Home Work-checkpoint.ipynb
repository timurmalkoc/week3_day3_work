### Pokemon API Wrapper
Create a python wrapper for the Pokemon API. It should take in a pokemon name and display the pokemon with its height and weight

```python
import requests
class PokemonAPI:
    def __init__(self):
        self.base_url = "https://pokeapi.co/api/v2/"
        
    def _get(self, api_method, query_param):
        request_url = f'{self.base_url}{api_method}{query_param.lower()}'
        pk_data = requests.get(request_url)
        print(request_url)
        if pk_data.ok:
            return pk_data.json()
        else:
            print("Something went wrong")
            
    def get_pokemon_info(self, name):
        pk_data = self._get('pokemon/', name)
        weight = pk_data["weight"]
        height = pk_data["height"]
        print(f"============ {pk_data['name']}'s info ===========")
        print(f'weight = {weight}\nheight = {height}')
        
    
    def get_pokemon_by_type(self, typ):
        pk_data = self._get('type/', typ)
        print(f"============ type {pk_data['name']} ===========")
        for pk in pk_data["pokemon"]:
            print(pk["pokemon"]["name"])
            
```

```python
pk = PokemonAPI()
pk.get_pokemon_info("bulbasaur")
pk.get_pokemon_info("Venusaur")
pk.get_pokemon_info("5")
```

```python
pk.get_pokemon_by_type("normal")
```

```python
pk.get_pokemon_by_type("4")
```
